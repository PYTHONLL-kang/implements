```python
import matplotlib.pyplot as plt
```

```python
def f(p, y):
    return (p**y) * (1-p)**(1-y)

def l(p, y_list):
    s = 1
    for y in y_list:
        s *= f(p, y)

    return s
```

```python
p_list = [0.01 * i for i in range(1, 100)]
y_list = [1, 1, 1, 0, 1, 1, 1, 0, 1, 0]
```

```python
plt.plot(p_list, [l(p, y_list) for p in p_list])
plt.xlabel('$\\theta$')
plt.ylabel('likelihood')
plt.show()
```
