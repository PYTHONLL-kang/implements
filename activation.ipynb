```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def d_sigmoid(x):
    o = sigmoid(x)
    return o * (1-o)
```

```python
def relu(x):
    return np.maximum(x,0)

def d_relu(x):
    o = relu(x)
    return [1 if x[i]==o[i] else 0 for i in range(len(x))]
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x)+np.exp(-x))

def d_tanh(x):
    o = tanh(x)
    return 1-o**2
```

```python
def leaky_relu(x):
    return np.maximum(x, 0.1*x)

def d_leaky_relu(x):
    o = leaky_relu(x)
    return [1 if x[i]==o[i] else 0.1 for i in range(len(x))]
```

```python
x = np.linspace(-10, 10, 100)
```

```python
def plot(x, f, d_f, name):
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.plot(x, f(x))
    plt.title(name)

    plt.subplot(1,2,2)
    plt.plot(x, d_f(x))
    plt.title(f'derivative of {name}')

    plt.show()
```

```python
plot(x, relu, d_relu, "ReLU")
```

```python
plot(x, leaky_relu, d_leaky_relu, "Leaky_ReLU")
```

```python
plot(x, sigmoid, d_sigmoid, "Sigmoid")
```

```python
plot(x, tanh, d_tanh, "Hyperbolic Tangent")
```
